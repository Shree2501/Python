```python
def factorial(n):
  if n<0:
    print("negative number factorial is not possible")
    return None
  elif n==0:
    return 1
  else:
    fact=1
    for i in range(1,n+1):
      fact=fact*i
    return fact
n=int(input("Enter a number: "))
result=factorial(n)
print(result)
```

```python
def calculate_area(d1,d2,shape):
  if shape=="rectangle":
    return d1*d2
  elif shape=="triangle":
    return 1/2*d1*d2
  else:
    print("invalid shape")
    return None
shape=input("Enter the shape: ")
d1=int(input("Enter the dimention1: "))
d2=int(input("Enter the dimention2: "))
print("area = ",calculate_area(d1,d2,shape))
```

```python
#wap to implement a fxn which return prime no from 1 to n
def prime(n):
  for i in range(2,n+1):
    isPrime=True
    for j in range(2,int(i/2)+1):
      if i%j==0:
        isPrime=False
        break
    if isPrime:
      print(i)
n=int(input("Enter a number: "))
prime(n)
```

```python
#WAP to implement a function that checks a given number is an Armstrong number or not
def isArmstrong(n):
  len=0
  sum=0
  origin = n
  while(n!=0):
    len+=1
    n//=10
  n=origin
  while(n!=0):
    dig=n%10
    sum+=(dig**len)
    n//=10
  if (origin==sum):
    print("Armstrong")
  else:
    print("Not Armstrong")

n=int(input("Enter a number: "))
isArmstrong(n)
```

```python
#WAP to implement a recursive function that checks a given number is an Armstrong number or not

def Armstrong(n,pow):
  if n == 0:
    return 0
  r = n % 10
  return (r**pow) + Armstrong(n//10,pow)

num=input("Enter a number : ")
pow=len(num)
n=int(num)
temp=n
if temp==Armstrong(n,pow):
  print("Armstrong")
else:
  print("Not Armstrong")
```

```python
l=[1,2,3,4]
k=int(input("Enter an integer: "))
a=[]

def maxNum(l,k):
  for i in l:
    a.append(i*k)
  max=a[0]
  for i in range(1,len(a)):
    if max<a[i]:
      max=a[i]
  return max

maxNum(l,k)
```
