```python
class human:
  def __init__(self,n,o):   #dunder
    self.name=n
    self.occupation=o
  def doWork(self):
    if self.occupation == "tennis player":
      print(self.name,"plays tennis")
    elif self.occupation == "actor":
      print(self.name,"shoots film")
  def speaks(self):
    print(self.name,"says hello!")
tom=human("tom cruise","actor")
tom.doWork()
tom.speaks()
maria=human("maria sharapowa","tennis player")
maria.doWork()
maria.speaks()
```

```python
from math import sqrt
class calc:
  def __init__(self,x):
    self.num = x
  def square(self):
    print(self.num**2)
  def cube(self):
    print(self.num**3)
  def squareRoot(self):
    print(sqrt(self.num))

value = calc(4)
value.square()
value.cube()
value.squareRoot()
```

```python
class rectangle:
  def __init__(self,l,b):
    self.length=l
    self.breadth=b
  def area(self):
    print("Area of rectangle: ",self.length*self.breadth)
  def perimeter(self):
    print("Perimeter : ",2*(self.length+self.breadth))

a=rectangle(2,3)
a.area()
a.perimeter()
```

```python
class bankAcc:
  def __init__(self,name,balance=0):
    self.name=name
    self.balance=balance
  def deposit(self,amount):
    self.balance+=amount
  def withdraw(self,amount):
    self.balance-=amount
  def showBal(self):
    print("Account holder:",self.name)
    print("Balance:",self.balance)

b=bankAcc("Shree",1000)
b.deposit(500)
b.withdraw(100)
b.showBal()
```

```python
class circle:
  def __init__(self,r):
    self.radius=r
  def area(self):
    print("Area of circle: ",3.14*self.radius*self.radius)
  def circumference(self):
    print("Perimeter : ",2*3.14*self.radius)

a=circle(3)
a.area()
a.circumference()
```

```python
class movie:
  def __init__(self,name,box):
    self.name=name
    self.box=box
  def hit_or_flop(self):
    if(self.box>100):
      print(self.name,"is hit")
    else:
      print(self.name,"is flop")
c=movie("I",30)
c.hit_or_flop()
b=movie("Sholay",200)
b.hit_or_flop()
```

```python
class cart:
  def __init__(self):
    self.item=[]
  def add_item(self,name,price):
    self.item.append((name,price))
  def show_total(self):
    return sum(price for name,price in self.item)

g=cart()
g.add_item("Soap",500)
g.add_item("brush",300)
print("Total price:",g.show_total())
```
