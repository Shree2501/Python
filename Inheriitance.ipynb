# **single_inheritance**

```python
class employee:
  company="ITC"

  def __init__(self,name,salary):
    self.name=name
    self.salary=salary

  def getInfo(self):
    print(f"The name is {self.name} and salary is {self.salary}")

class programmer(employee):
  company="ITC Infotech"

a=employee("Shree",1200000)
b=programmer("Rohan",120000)
print(a.company,b.company)
a.getInfo()
b.getInfo()
```

```python
#multiple inheritance
class employee:
  company="ITC"

  def __init__(self,name,salary):
    self.name=name
    self.salary=salary

  def getInfo(self):
    print(f"The name is {self.name} and salary is {self.salary}")

class coder:
  language="Python"
  def display(self):
    print(f"The language is {self.language}")

class programmer(employee,coder):
  language="C"
  company="ITC Infotech"

a=employee("Shree",1200000)
b=programmer("Rohan",120000)
c=coder()
print(a.company,b.company)
a.getInfo()
b.getInfo()
b.display()
```

```python
#multi level inheritance
class employee:
  company="ITC"

  def __init__(self,name,salary):
    self.name=name
    self.salary=salary

  def getInfo(self):
    print(f"The name is {self.name} and salary is {self.salary}")

class coder(employee):
  language="Python"
  def display(self):
    print(f"The language is {self.language}")

class programmer(coder):
  language="C"
  company="ITC Infotech"

a=employee("Shree",1200000)
b=programmer("Rohan",120000)
c=coder("Anu",12)
print(a.company,b.company)
a.getInfo()
b.getInfo()
b.display()
c.getInfo()
```

SPECIAL METHODS


```python
class employee:
  name="Anu"
  @classmethod
  def getInfo(cls):
    print(f"The name is {cls.name}")

a=employee()
a.name="Shree"
a.getInfo()
```

```python
class employee:
  age=18

  @classmethod
  def getInfo(cls):
    print(f"The Age is {cls.age}")

  @property
  def name(self):
    return f"{self.fname} {self.lname}"

  @name.setter
  def name(self,value):
    self.fname=value.split(" ")[0]
    self.lname=value.split(" ")[1]


e=employee()
e.getInfo()

e.name="Shree Sharma"
print(e.name)
```

```python
class twoDvector:
  def __init__(self,i,j):
    self.i=i
    self.j=j

  def show(self):
    print(f"{self.i}i {self.j}j")
class threeDvector(twoDvector):
  def __init__(self,i,j,k):
    super().__init__(i,j)
    self.k=k
  def show(self):
    print(f"{self.i}i {self.j}j {self.k}k")

a=twoDvector(1,2)
b=threeDvector(6,2,3)
a.show()
b.show()
```

```python
class Complex:
  def __init__(self,r,i):
    self.r=r
    self.i=i
  def __add__(self,a):
    return Complex(self.r + a.r,self.i + a.i)
  def __str__(self):
    return f"{self.r} + {self.i}i"
c1 = Complex(1,2)
c2 = Complex(3,4)
print(c1 + c2)
```

```python
class vector:
  def __init__(self,i,j,k):
    self.i=i
    self.j=j
    self.k=k
  def __add__(self,u):
    return vector(self.i + u.i, self.j + u.j, self.k + u.k)

  def __mul__(self,u):
    return self.i * u.i + self.j * u.j + self.k * u.k

  def __str__(self):
    return f"{self.i}i + {self.j}j + {self.k}k"

v1=vector(1,2,3)
v2=vector(4,5,6)
print(v1+v2)
print(v1*v2)
```

```python
class vector:
  def __init__(self,l):
    self.l=l

  def __len__(self):
    return len(self.l)

v1 = vector([4,5,6])
print(len(v1))
```

```python
class grandfather:
  def __init__(self,grandfathername):
    self.grandfathername=grandfathername

class father(grandfather):
  def __init__(self,fathername,grandfathername):
    self.fathername=fathername
    grandfather.__init__(self,grandfathername)

class son(father):
  def __init__(self,sonname,fathername,grandfathername):
      self.sonname=sonname
      father.__init__(self,fathername,grandfathername)

  def print_name(self):
    print("Grnadfather name: ",self.grandfathername)
    print("Father name: ",self.fathername)
    print("Son name: ",self.sonname)

s1=son("Prince","Rampal","Lal mani")
print(s1.grandfathername)
s1.print_name()
```

```python
class Person:
  def __init__(self,name,age):
    self.name=name
    self.age=age
class Employee(Person):
  def __init__(self,name,age,eid,salary):
    super().__init__(name,age)
    self.eid=eid
    self.salary=salary
  def display(self):
    print("Name:",self.name)
    print("Age:",self.age)
    print("E_id:",self.eid)
    print("Salary:",self.salary)

e=Employee("Shree",18,"e1",200000000000)
e.display()
```

```python
class student:
  def __init__(self,rollno,name):
    self.rollno=rollno
    self.name=name

class marks(student):
  def __init__(self,m1,m2,m3,rollno,name):
    self.m1=m1
    self.m2=m2
    self.m3=m3
    student.__init__(self,rollno,name)
  def display(self):
    print("name:",self.name)
    print("rollno:",self.rollno)
    print("Total:",self.m1+self.m2+self.m3)
    print("Percentage:",(self.m1+self.m2+self.m3)//3)

s=marks(100,86,98,60,"Shree")
s.display()
```
