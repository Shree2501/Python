```python
num = [2, 3, 1, 4, 6, 2]
num.sort() # Binary search requires a sorted list
print("Sorted list:", num)

a = int(input("Enter a no. to search: "))

low = 0
high = len(num) - 1
found_index = -1 # Initialize to -1 to indicate element not found

while low <= high:
    mid = (low + high) // 2
    mid_val = num[mid]

    if mid_val == a:
        found_index = mid
        break  # Element found, exit the loop
    elif mid_val < a:
        low = mid + 1 # Search in the right half
    else:
        high = mid - 1 # Search in the left half

if found_index != -1:
    print(f"Element found at index: {found_index}")
else:
    print("Element not found")
```

```python
a=[[1,2,3],[],[4,5,6]]
r_sum=[]
for i in range(len(a)):
  sum=0
  for j in range(len(a[i])):
    sum+=a[i][j]
  if(sum>0):
    r_sum.append(sum)
print(r_sum)
```
