# **ENCAPSULATION**

```python
class Computer:
  def __init__(self):
    self.maxprice=900
  def sell(self):
    print("Selling Price: {}".format(self.maxprice))
  def setMaxPrice(self,price):
    self.maxprice=price

c=Computer()
c.sell()
c.maxprice=1000
c.sell()
c.setMaxPrice(1000)
c.sell()
```

# **OVERLOADING**

```python
class polygon:
  def render(self):
    print("Rendering polygon...")
class square(polygon):
  def render(self):
    print("Rendering square...")
class circle(polygon):
  def render(self):
    print("Rendering circle...")

s1=square()
s1.render()
c1=circle()
c1.render()
```

```python
class MethodOverloading:
  def method(self):
    print("Method without argument")
  def method(self,x):
    print("Method with single argument")
  def method(self,x,y):
    print("Method with two arguments")

obj= MethodOverloading()
obj.method()
obj.method(2)
obj.method(3,4)
```
