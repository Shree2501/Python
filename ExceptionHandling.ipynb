```python
def grade(marks:int):
  if 90<=marks<=100:
    return 'A'
  elif 80<=marks<90:
    return 'B'
  elif 70<=marks<80:
    return 'C'
  elif 60<=marks<70:
    return 'D'
  elif 0<=marks<60:
    return 'Fail'
  else:
    return None

print("Grade = ",grade)
try:
    marks=int(input("Enter your marks: "))
    if marks > 100 or marks < 0:
      raise ValueError("Grade must be in range 0-100")
    letter=grade(marks)
except ValueError as ve:
  print(f"{ve}")
else:
  print(f"Your grade is: {letter}")
finally:
  print("Thanks for using our claculator")
```
