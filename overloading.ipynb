```python
class vehicle:
  def speed(self):
    print("Vehicle speed is 60km/h")
class car(vehicle):
  def speed(self):
    super().speed()
    print("Car speed is 120km/h")

v=vehicle()
c=car()
v.speed()
c.speed()
```

```python
class shape:
  def area(self):
    print("Calculating area")
class circle(shape):
  def area(self,r):
    print("circle area:",3.14*r*r)
class rectangle(shape):
  def area(self,l,b):
    print("rectangke area:",l*b)
s=shape()
s.area()
c=circle()
c.area(4)
r=rectangle()
r.area(3,2)
```
